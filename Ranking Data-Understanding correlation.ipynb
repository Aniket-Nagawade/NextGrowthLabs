```python
import re
import pandas as pd
import numpy as np
import string
```

```python
df = pd.read_csv(r'D:\NextLabs\browser_rankings_data.csv')
df.head()
```

```python
df.shape
```

```python
print("Null data in the train dataset:\n", df.isnull().any())
```

```python
df[df['Rank'].isnull()]
```

```python
df = df.dropna()
df.shape
```

```python
df['Short Desc Len'] = df['Short Description'].apply(lambda x:len(x))
df['Long Desc Len'] = df['Long Description'].apply(lambda x:len(x))
```

```python
Key_list=[]
for text in df['Long Description']:
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub("\s+", " ", text)
    text = re.sub('[^a-zA-Z0-9\n]', ' ', text)
    #text = re.sub("\t", " ", text)  
    text = text.split(" ")
    if 'Key' in text[0:50]:
        Key_list.append(1)
    else:
        Key_list.append(0)
```

```python
df['Key_Present_in50'] = Key_list
```

```python
df.head(2)
```

```python
df['App ID'].unique()
```

```python
some_map = {label:idx for idx,label in enumerate(np.unique(df['App ID']))} #create a map w.r.t the App Id column
df['App_id_enc'] = df['App ID'].map(some_map ) #assign values to App_id_enc column
```

```python
Keyword_CrosstabResult=pd.crosstab(index=df['Keyword'],columns=df['Rank'])
print(Keyword_CrosstabResult)
```

```python
df['Keyword'].unique()
```

```python
some_map = {label:idx for idx,label in enumerate(np.unique(df['Keyword']))} #create a map w.r.t the Keyword column
df['Keyword_enc'] = df['Keyword'].map(some_map ) #assign values to Keyword enc column
```

```python
Keyword_CrosstabResult=pd.crosstab(index=df['Keyword'],columns=df['Rank'])
print(Keyword_CrosstabResult)
```
